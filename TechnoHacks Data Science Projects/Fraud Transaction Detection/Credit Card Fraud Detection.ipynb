```python
#Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Reading the dataset
df=pd.read_csv("creditcard.csv")
df.head()
```

```python
#Getting more info of the dataset
df.describe()
```

```python
#Checking for null values
df.isna().sum()
```

```python
#Distributing Legit Transactions and Fraudlent Transactions
df["Class"].value_counts()
```

```python
#Separating the dataset for analysis
df1 = df[df.Class == 0] #Legit Transactions
print(df1.shape)
df2 = df[df.Class == 1] #Fraud Transactions
print(df2.shape)
```

```python
#Retrieving amount information for both datasets
print(df1.Amount.describe())
print()
print(df2.Amount.describe())
print()
```

```python
#Comparing the values
df.groupby("Class").mean()
```

```python
#Building a dataset with similar distribution of legit and fraud transactions
df3 = df1.sample(n=492) 
df4 = pd.concat([df3,df2], axis = 0)
df4.head()
```

```python
df4.tail()
```

```python
#Checking the distribution of new dataset
df4["Class"].value_counts()
```

```python
#Comparing the values of new dataset
df4.groupby("Class").mean()
```

```python
#Model Training
X = df4.drop(columns = "Class", axis = 1)
Y = df4["Class"]

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size = 0.8, stratify = Y, random_state = 2)

model=LogisticRegression(random_state=0)
model.fit(X_train, Y_train)
```

```python
#Model Prediction
pred = model.predict(X_test)
pred
```

```python
print(Y_test)
```

```python
#Checking accuracy of the model
accuracy = accuracy_score(pred,Y_test)
accuracy
```
